```python
import numpy as np
import pandas as pd
import string
import nltk
import math
nltk.download('stopwords')
```

```python
corpusA =  np.genfromtxt("/content/sample_data/windows.txt", dtype=str, encoding='utf-8-sig')
corpusB = np.genfromtxt("/content/sample_data/linux.txt", dtype=str, encoding='utf-8-sig')
corpusC =  np.genfromtxt("/content/sample_data/mac.txt", dtype=str, encoding='utf-8-sig')
```

```python
corpusA
```

```python
def remove_punctuation(text):
  punctuationFree = "".join([i for i in text if i not in string.punctuation])
  return punctuationFree
```

```python
def remove_stopwords(text):
  stopwords = nltk.corpus.stopwords.words('english')
  stopwordsFree = [i for i in text if i not in stopwords]
  return stopwordsFree
```

```python
def preprocessing(text):
  punctuationFreeText = []
  for word in text:
    lowerCaseWord = word.lower()
    punctuationFreeText.append(remove_punctuation(lowerCaseWord))

  stopWordsFreeText = remove_stopwords(punctuationFreeText)
  return stopWordsFreeText
```

```python
corpusA = np.array(preprocessing(corpusA))
corpusB = np.array(preprocessing(corpusB))
corpusC = np.array(preprocessing(corpusC))
```

```python
corpusA
```

```python
uniqueWords = set(corpusA).union(set(corpusB).union(corpusC))
```

```python
def countWordOccurrence(uniqueWords, bagOfWords):
  wordOccurence = dict.fromkeys(uniqueWords, 0)
  for word in bagOfWords:
    wordOccurence[word] += 1

  return wordOccurence
```

```python
wordOccurenceA = countWordOccurrence(uniqueWords, corpusA)
wordOccurenceB = countWordOccurrence(uniqueWords, corpusB)
wordOccurenceC = countWordOccurrence(uniqueWords, corpusC)
```

```python
wordOccurenceA
```

```python
def computeTF(wordDict, bagOfWords):
    tfDict = {}
    bagOfWordsCount = len(bagOfWords)
    for word, count in wordDict.items():
        tfDict[word] = count / float(bagOfWordsCount)
    return tfDict
```

```python
TFdocA = computeTF(wordOccurenceA, corpusA)
TFdocB = computeTF(wordOccurenceB, corpusB)
TFdocC = computeTF(wordOccurenceC, corpusC)
```

```python
TFdocA
```

```python
def computeIDF(documents):
  N = len(documents)
  idfDict = dict.fromkeys(documents[0].keys() , 0)
  for document in documents:
    for word, val in document.items():
      if val > 0:
        if word in idfDict.keys():
          idfDict[word] += 1
        else:
          idfDict[word] = 1

  for word, val in idfDict.items():
    idfDict[word] = math.log(N / float(val))
  return idfDict
```

```python
idfDict = computeIDF([wordOccurenceA, wordOccurenceB, wordOccurenceC])
```

```python
idfDict
```

```python
def computeTFIDF(tfBagOfWords, idfs):
    tfidf = {}
    for word, val in tfBagOfWords.items():
        tfidf[word] = val * idfs[word]
    return tfidf
```

```python
TFIDF_A = computeTFIDF(TFdocA, idfDict) 
TFIDF_B = computeTFIDF(TFdocB, idfDict)
TFIDF_C = computeTFIDF(TFdocC, idfDict) 
```

```python
df = pd.DataFrame([TFIDF_A, TFIDF_B, TFIDF_C])
```

```python
def euclidean_dist(a,b):
  a = np.array(list(a.values()))
  b = np.array(list(b.values()))
  dist = np.sqrt(np.sum(np.square(a-b)))
  return dist
```

```python
print(euclidean_dist(TFIDF_A, TFIDF_C))
print(euclidean_dist(TFIDF_B, TFIDF_C))
```
